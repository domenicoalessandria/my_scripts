```python
import random
import time
from IPython.display import clear_output
```

```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
```

```python
class Card():
    
    def __init__(self,suit,rank):
        
        self.suit=suit
        self.rank=rank
        self.value=values[rank]
        
    def __str__(self):
        
        return self.rank + ' of '+ self.suit
```

```python
class Deck():

    
    def __init__(self):
        
     
        self.all_cards=[]
        
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))
                
    def __str__(self):
        
        deck_str=''
        for card in self.all_cards:
            deck_str+= '\n'+ card.__str__()
        return deck_str
    
    def shuffle(self):
        
        random.shuffle(self.all_cards)
    
    def deal_one(self):
        
        return self.all_cards.pop(0)
    
    def restore(self):
        
        self.all_cards=[]
        
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))
        
        
        
```

```python
class Player():
    
    def __init__(self,name):
        
        self.name=name
        self.funds=500
        self.own_cards=[]
        self.hand=''
        self.score=0
        
        
    def score_f(self):
        
        self.score=0
        for item in self.own_cards:
            self.score+=item.value
        
        if self.score>21:
            for card in self.own_cards:
                if card.rank=='Ace':
                    self.score-=10
                if self.score<=21:
                    break
            
        return self.score
    
    
    def __str__(self):
        
        self.hand=''
        for card in self.own_cards:
            self.hand+= '\n' + card.__str__()
        return self.hand
            
    
    def clear(self):
        
        self.own_cards=[]
                           
```

```python
class Dealer():
    
    def __init__(self):
        
        self.own_cards=[]
        self.hand=''
        self.score=0
        
    def score_func(self):
        
        self.score=0
        for item in self.own_cards:
            self.score+=item.value
        
        if self.score>21:
            for card in self.own_cards:
                if card.rank=='Ace':
                    self.score-=10
                if self.score<=21:
                    break
                    
        return self.score
    
    
    def __str__(self):
        self.hand=''
        for card in self.own_cards:
            self.hand+= '\n' + card.__str__()
        return self.hand

        
    def clear(self):
        
        self.own_cards=[]
                           
```

```python
## Do you want to play?
game_on_input=''

while game_on_input!='y' and game_on_input!='n':

    game_on_input=input("\nDo you want to play Blackjack? Type 'y'(yes) or 'n'(no): ").lower()

    if game_on_input=='y':
        game_on=True
    elif game_on_input=='n':
        game_on=False
    else:
        print('Invalid selection \nTry again')
        

##Asking how many players there are (max 3)

if game_on:
    how_many_player=''

    while how_many_player not in (1,2,3):
        try:
            how_many_player=int(input('How many player? (Max 3) Insert a number between 1 and 3: '))
        except:
            print("Insert an anctual number (eg:'1')")

        if how_many_player==1:
            one_play=True
            two_play=False
            three_play=False
        elif how_many_player==2:
            one_play=True
            two_play=True
            three_play=False
        elif how_many_player==3:
            one_play=True
            two_play=True
            three_play=True
        else:
            print('Invalid selection')

##Creating the variables that store the number of games in the current session
game_num=0

##Starting the game
while game_on:
    
##Clearing the output for the next round of the same session   
    clear_output()
    
    game_num+=1 
    print(f'Game number {game_num}')
    
##Initializing the players
    if game_num==1:
        deck=Deck()
        dealer=Dealer()
        
        if one_play:
            one_name= input("Type player 1 name: ")
            one=Player(one_name)

        if two_play:
            two_name= input("Type player 2 name: ")
            two=Player(two_name)

        if three_play:
            three_name= input("Type player 3 name: ")
            three=Player(three_name)
        
    else:
        deck.restore()
        dealer.clear()
        if one_play:
            one.clear()
        if two_play:
            two.clear()
        if three_play:
            three.clear()

##Shuffling the deck

    deck.shuffle()

## Checking if the players have any chips and eventually how much of them they want to bet 

    if one_play and one.funds>0:
        
        print(f'{one.name} has {one.funds} chips')
        
        one_bet=one.funds+1
        
        while one_bet>one.funds:
            one_bet=int(input (f'{one.name}, digit the amount of chips you want to bet as a number: '))

            if one.funds>=one_bet:
                continue
            else:
                print('Not enough chips \nTry again')

        print(f'{one.name} bet {one_bet} chips')
    else:
        one_play=False


    if two_play and two.funds>0:
        
        print(f'{two.name} has {two.funds} chips')

        two_bet=two.funds+1

        while two_bet>two.funds:
            two_bet=int(input (f'{two.name}, digit the amount of chips you want to bet as a number: '))

            if two.funds>=two_bet:
                continue
            else:
                print('Not enough chips \nTry again')

        print(f'{two.name} bet {two_bet} chips')
    else:
        two_play=False
        
        
    if three_play and three.funds>0:
        
        print(f'{three.name} has {three.funds} chips')

        three_bet=three.funds+1

        while three_bet>three.funds:
            three_bet=int(input (f'{three.name}, digit the amount of chips you want to bet as a number: '))

            if three.funds>=three_bet:
                continue
            else:
                print('Not enough chips \nTry again')

        print(f'{three.name} bet {three_bet} chips')
    else:
        three_play=False

    
##Dealing the Players' cards

    time.sleep(1)
    dealer.own_cards.append(deck.deal_one())
    dealer.own_cards.append(deck.deal_one())
    print(f'\nDealer hand: \n<<hidden card>> \n{dealer.own_cards[1]}')

    if one_play:
        time.sleep(1)
        one.own_cards.append(deck.deal_one())
        one.own_cards.append(deck.deal_one())
        one.score_f()
        print(f'{one}\n{one.name} has {one.score} points')


    if two_play:
        time.sleep(1)
        two.own_cards.append(deck.deal_one())
        two.own_cards.append(deck.deal_one())
        two.score_f()
        print(f'{two}\n{two.name} has {two.score} points')


    if three_play:
        time.sleep(1)
        three.own_cards.append(deck.deal_one())
        three.own_cards.append(deck.deal_one())
        three.score_f()
        print(f'{three}\n{three.name} has {three.score} points')


## Players' choices

    print("\nBeginning of players' choices")

    if one_play:
        player_choice=''
        one_turn= True
        print(f'\n{one.name} turn')
        one_busted=False

        while one_turn:
            
            if one.score!=21:
                
                while player_choice!='hit' and player_choice!='stand':

                    player_choice=input(f"Please {one.name} type 'hit' or 'stand': ").lower()

                    if player_choice!='hit' and player_choice!='stand': 
                        print('Invalid selection')

                if player_choice=='hit':
                    player_choice=''
                    one.own_cards.append(deck.deal_one())
                    one.score_f()
                    print(f'{one.name} gets {one.own_cards[-1]} \n{one.name} has now {one.score} points')

                    if one.score>21:
                        print(f'{one.name} busts!')
                        one_turn=False
                        one_busted=True
                        one.funds-=one_bet
                        
                    elif one.score==21:
                        one_turn=False
                        print(f'{one.name} gets Blackjack')

                if player_choice=='stand':
                    one_turn=False
                    time.sleep(0.5)
                    
            else:  
                one_turn=False
                print(f'\n{one.name} gets Blackjack!')

    
    if two_play:
        player_choice=''
        two_turn= True
        print(f'\n{two.name} turn')
        two_busted=False

        while two_turn:
            
            if two.score!=21:
                
                while player_choice!='hit' and player_choice!='stand':

                    player_choice=input(f"Please {two.name} type 'hit' or 'stand': ").lower()

                    if player_choice!='hit' and player_choice!='stand': 
                        print('Invalid selection')

                if player_choice=='hit':
                    player_choice=''
                    two.own_cards.append(deck.deal_one())
                    two.score_f()
                    print(f'{two.name} gets {two.own_cards[-1]} \n{two.name} has now {two.score} points')

                    if two.score>21:
                        print(f'{two.name} busts!')
                        two_turn=False
                        two_busted=True
                        two.funds-=one_bet
                        
                    elif two.score==21:
                        two_turn=False
                        print(f'{two.name} gets Blackjack')

                if player_choice=='stand':
                    two_turn=False
                    time.sleep(0.5)
                    
            else:  
                two_turn=False
                print(f'{two.name} gets Blackjack!')

                
    if three_play:
        player_choice=''
        three_turn= True
        print(f'\n{three.name} turn')
        three_busted=False

        while three_turn:
            
            if three.score!=21:
                
                while player_choice!='hit' and player_choice!='stand':

                    player_choice=input(f"Please {three.name} type 'hit' or 'stand': ").lower()

                    if player_choice!='hit' and player_choice!='stand': 
                        print('Invalid selection')

                if player_choice=='hit':
                    player_choice=''
                    three.own_cards.append(deck.deal_one())
                    three.score_f()
                    print(f'\n{three.name} gets {three.own_cards[-1]} \n{three.name} has now {three.score} points')

                    if three.score>21:
                        print(f'{three.name} busts!')
                        three_turn=False
                        three_busted=True
                        three.funds-=one_bet
                        
                    elif three.score==21:
                        three_turn=False
                        print(f'{three.name} gets Blackjack')

                if player_choice=='stand':
                    three_turn=False
                    time.sleep(0.5)
                    
            else:  
                three_turn=False
                print(f'{three.name} gets Blackjack!')
                
    print('\nEnd of Players turns.')
    
    time.sleep(0.4)
    
##Dealer game

    dealer.score_func()
    dealer_busted=False
    print(f'\nThe Dealer reveals his covered card: \n{dealer.own_cards[0]} \n{dealer.own_cards[1]}\nDealer has {dealer.score} points')
    
    while dealer.score<17:
        dealer.own_cards.append(deck.deal_one())
        dealer.score_func()
        print(f'\nThe Dealer draws: \n{dealer.own_cards[-1]} \nDealer has now {dealer.score} points')

        if dealer.score>21:
            print(f'\nThe Dealer busts!')
            dealer_busted=True
            
        time.sleep(0.5)
        
    if dealer.score==21:
        print('\nThe Dealer gets Blackjack')

    time.sleep(1.3)
    
    if one_play:
        if dealer_busted and not one_busted:
            print(f'{one.name} wins')
            one.funds+=one_bet

        if not one_busted and not dealer_busted:
            if dealer.score>one.score:
                print(f'\nDealer wins against {one.name}')
                one.funds-=one_bet
            elif dealer.score<one.score:
                print(f'\n{one.name} wins!')
                one.funds+=one_bet
            else:
                if one.score==21 and dealer.score==21:
                    print(f'\n{one.name} push')
                else:
                    print(f'\n{one.name} tie')
   

    time.sleep(0.5)
    
    if two_play:
        if dealer_busted and not two_busted:
            print(f'{two.name} wins')
            two.funds+=two_bet
            
        if not two_busted and not dealer_busted:
            if dealer.score>two.score:
                print(f'\nDealer wins against {two.name}')
                two.funds-=two_bet
            elif dealer.score<two.score:
                print(f'\n{two.name} wins!')
                two.funds+=two_bet
            else:
                if two.score==21 and dealer.score==21:
                    print(f'\n{two.name} push')
                else:
                    print(f'\n{two.name} tie')
    
        
    time.sleep(0.5)
    if three_play:
        if dealer_busted and not three_busted:
            print(f'{three.name} wins')
            three.funds+=three_bet

        if not three_busted and not dealer_busted:
            if dealer.score>three.score:
                print(f'\nDealer wins against {three.name}')
                three.funds-=three_bet
            elif dealer.score<three.score:
                print(f'\n{three.name} wins!')
                three.funds+=three_bet
            else:
                if three.score==21 and dealer.score==21:
                    print(f'\n{three.name} push')
                else:
                    print(f'\n{three.name} tie')
    
##Do you want to play again?

    time.sleep(5)
    closer=True           
    while closer:
        play_again= input("Do you wanto to play again? Type 'y'(yes) or 'n'(no) ")

        if play_again=='y':
            closer= False
        elif play_again=='n':
            game_on= False
            closer= False
        else:
            print('Invalid Selection')

     
```
