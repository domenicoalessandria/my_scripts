```python
def player_input(choice):
    
    while choice!= 'X' and choice!= 'O':
        
        choice=input('Player 1, please select X or O: ')
        
        if choice=='X' or choice=='O':
            if choice== 'X':
                choice_2= 'O'
            elif choice=='O':
                choice_2='X'
    
            global P1_marker
        
            global P2_marker
                
            P1_marker= choice
            P2_marker= choice_2
            
            return (print(f'Player 1 chose: {choice} \nPlayer 2 is assigned: {choice_2}'))
            
        else:
            print('Sorry, you need to enter either X or O!')
```

```python
who=1
def place_marker(board):
    
    global location_choosen
    
    while_loop_index=1
    
    location_choosen='wrong'
    
    while not location_choosen.isdigit() or while_loop_index==1:
        
        location_choosen=input('Please select a number between 1 and 9: ')
            
        if not location_choosen.isdigit():
            print('Please digit an actual number')
            
        elif int(location_choosen) not in range(1,10):
            print('Selected number out of range')
            
        elif board[int(location_choosen)]!=' ':
            print('Please select an empty position')
        
        else:
            global who
            if who==1:
                board[int(location_choosen)]='X'
                who=2
                while_loop_index=0
            else:
                board[int(location_choosen)]='O'
                who=1
                while_loop_index=0
            
```
