```python
def fatture_detec(mylist,the_sum):
    for i in range(2,len(mylist)):
        c=comb(mylist,i)
        c_list=list(c)
        for i in range(len(c_list)):
            t_list=list(c_list[i])
            my_sum=0
            for i in range(len(t_list)):
                my_sum+=t_list[i]
            if my_sum<0.5+the_sum and my_sum>the_sum-0.5:
                print(my_sum)
                print(t_list)
```
